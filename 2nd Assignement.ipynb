```python
#4 Compare the Triplets

import math
import os
import random
import re
import sys

def compareTriplets(a, b):
    alice = 0
    bob = 0
    for i in range(len(a)):
        if a[i] > b[i]:
            alice = alice + 1
        elif a[i] < b[i]:
            bob = bob + 1  
        else:
            alice = alice
            bob = bob
    return [alice, bob]

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
#5 Diagonal difference

import math
import os
import random
import re
import sys

def diagonalDifference(arr):
    ltr = 0
    rtl = 0
    diff = 0
    for i in range(len(arr)):
        ltr += arr[i][i]
        rtl += arr[i][len(arr)-i -1]
        diff = ltr - rtl
        if diff < 0:
            diff = diff * -1
    return diff

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#6 Staircase

import math
import os
import random
import re
import sys

def staircase(n):
    for i in range(n):
        for j in range(n):
            if (i+j)>=(n-1):
                print('#',end="")
            else:
                print(end=" ")
        print(end='\n')

if __name__ == '__main__':
    n = int(input().strip())

    staircase(n)
```

```python
#7 Plus Minus

import math
import os
import random
import re
import sys

def plusMinus(arr):
    n = len(arr)
    pos = 0 
    neg = 0
    zero = 0
    for i in arr:
        if i < 0:
            neg += 1
        elif i == 0:
            zero += 1
        elif i > 0:
            pos += 1
    print(round(pos/n, 6),round(neg/n, 6),round(zero/n, 6),sep="\n")
    
if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    plusMinus(arr)
```

```python
#9 Time Conversion

import math
import os
import random
import re
import sys

def timeConversion(s):
    hour = s[0:2]
    minutes = s[3:5]
    seconds = s[6:8]
    am_pm = s[-2:]
    if hour == "12" and am_pm == "PM":
        hour = "12"
    elif hour == "12" and am_pm == "AM":
        hour = "00"
    elif am_pm == "PM":
        hour = int(hour) + 12
    military = str(hour) + ":" + minutes + ":" + seconds
    return military

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#2 Quicksort1 - Partition

import math
import os
import random
import re
import sys

def quickSort(arr):
    left = []
    right = []
    equal = []
    for p in arr:
        if p == arr[0]:
            equal.append(p)
        elif p < arr[0]:
            left.append(p)
        else:
            right.append(p)        
    arr = left + equal + right
    return arr

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = quickSort(arr)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
#3 Find the Median

import math
import os
import random
import re
import sys

def findMedian(a):
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and key < a[j]:
            a[j+1] = a[j]
            j -= 1
            a[j+1] = key
    return a[len(a)//2]

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = findMedian(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#6 Running Time of Algorithms

import math
import os
import random
import re
import sys

def runningTime(arr):
    temp = 0
    for i in range(1, len(arr)):
        j = i-1
        key = arr[i]
        while j >= 0 and arr[j] > key:
           arr[j+1] = arr[j]
           j -= 1
           temp += 1
        arr[j+1] = key
    return temp

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = runningTime(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#2 Mars Exploration

import math
import os
import random
import re
import sys

def marsExploration(s):
    msg = 'SOS'
    errors = 0
    for i in range(len(s)):
        s_3 = msg[(i%3)]
        if s_3 != s[i]:
            errors += 1
    return errors

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = marsExploration(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#3 CamelCase

import math
import os
import random
import re
import sys


def camelcase(s):
    words = 1
    for i in s:
        if i.isupper():
            words = words + 1
    return words
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = camelcase(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#4 Alternating Charachters

import math
import os
import random
import re
import sys

def alternatingCharacters(s):
    result = 0
    for i in range(1,len(s)):
        if s[i] == s[i-1]:
            result += 1
        else:
            result += 0
    return result

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        s = input()

        result = alternatingCharacters(s)

        fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#5 Game of thrones

import math
import os
import random
import re
import sys

def gameOfThrones(s):
    from collections import Counter
    temp = Counter(s)
    count = 0 
    for i in temp:
        if (temp[i]%2) != 0:
            count+=1 
    if count == 0 or count == 1:
        return 'YES'
    else:
        return 'NO'
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = gameOfThrones(s)

    fptr.write(result + '\n')

    fptr.close()
```
